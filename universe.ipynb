```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
names = []
def get_names():
    with open('constellations/stars.txt', 'r') as f:
        for line in f:
            data = line.split()
            if len(data) > 6:
                nom = ""
                for word in data[6:]:
                    if word.endswith(';'):
                        nom += word.replace(';', '') + " "
                        break
                    else:
                        nom += str(word) + " "
                names.append(nom)
            else:
                names.append("")
    return names

concatenated_names = get_names()
print(concatenated_names)
```

```python
names = get_names()
data = np.loadtxt('constellations\stars.txt', delimiter=' ', skiprows=1, usecols=(0, 1, 2, 3))
x = data[:, 0]  # primera columna (x)
y = data[:, 1]  # segunda columna (y)
z = data[:, 2]  # tercera columna (z)
text_names = [str(name) for name in names]
```

```python
#Buscar las coordenadas para conectar las estrellas 
def constellation(txt_archivo, fig, op):
    with open(txt_archivo, 'r') as f:
        for line in f:
            data = line.split(",")
            star1 = data[0].replace("\n", "")
            star2 = data[1].replace("\n", "")
            try:
                index1 = names.index(str(star1 + " "))
                index2 = names.index(str(star2 + " "))
                if op == 1:
                    fig.add_trace(go.Scatter3d(
                    x=[x[index1], x[index2]],
                    y=[y[index1], y[index2]],
                    z=[z[index1], z[index2]],
                    mode='lines', line=dict(color='white')))
                elif op == 2:
                    fig.add_trace(go.Scatter(
                        x=[x[index1], x[index2]],
                        y=[y[index1], y[index2]],
                        mode='lines',
                        line=dict(color='white')))

            except ValueError:
                continue   
```

```python
def personalized_figure(fig):
    fig.update_layout(scene=dict(xaxis=dict(showbackground=False),
                                yaxis=dict(showbackground=False),
                                zaxis=dict(showbackground=False)),
                    plot_bgcolor='black',  # Fondo negro
                    paper_bgcolor='black',  # Fondo negro
                    font=dict(color='white'),width=1000,  # Ancho de la gráfica en píxeles
                    height=1000)  # Fuente blanca
```

```python
def brightness(x, vector):
    brillo_min = 0.1  # Valor mínimo del brillo
    brillo_max = 1.0  # Valor máximo del brillo
    brillo = np.random.uniform(brillo_min, brillo_max, len(x))

    # Normalizar los valores del brillo al rango [0, 1]
    brillo_normalized = (brillo - brillo_min) / (brillo_max - brillo_min)
    # Asignar colores basados en el brillo
    vector = ['rgb(255,255,255)' if b > 0.5 else 'rgb(150,150,150)' for b in brillo_normalized]
    return vector
```

```python
def draw_stars(x, y, z, text_names):
    # Generar valores aleatorios para el brillo
    colors = []
    colors = brightness(x, colors)
    fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, text=text_names, mode='markers',
                                    marker=dict(color=colors, size=1))])

    personalized_figure(fig)
    constellation("constellations\Osa mayor.txt",fig, 1)
    constellation("constellations\Osa menor.txt",fig, 1)
    constellation("constellations\Hydra.txt",fig, 1)
    constellation("constellations\Geminis.txt",fig, 1)
    constellation("constellations\Cygnet.txt",fig, 1)
    constellation("constellations\Cazo.txt",fig, 1)


    fig.show()
```

```python
draw_stars(x, y, z, text_names)
```
